```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.patches as patches
import seaborn as sns
import json
import numpy as np
from datetime import datetime
from IPython.display import display, HTML

# ==================== Konfigurasi Path ====================
base_path = "saved_models_arima"
gambar_path = os.path.join(base_path, "gambar")
os.makedirs(gambar_path, exist_ok=True)
sentiment_folder = "saved_sentiment_scraping"
os.makedirs(sentiment_folder, exist_ok=True)

# ==================== Fungsi untuk menampilkan tabel dengan 4 digit ====================
def display_table(df, title):
    # Format semua kolom float menjadi 4 digit
    styled_df = df.copy()
    float_cols = styled_df.select_dtypes(include=['float']).columns
    for col in float_cols:
        styled_df[col] = styled_df[col].apply(lambda x: f"{x:.4f}" if pd.notnull(x) else x)
    
    # Tampilkan tabel dengan gaya
    display(HTML(f"<h3>{title}</h3>"))
    display(styled_df)
    print("\n")

# ==================== Load Data ====================
raw_data_path = os.path.join(base_path, "raw_stock_prices_aapl.csv")
sentiment_path = os.path.join(sentiment_folder, "sentiment_aapl.csv")
processed_features_path = os.path.join(base_path, "processed_features_aapl.csv")

# Load data saham dengan format yang benar
data_raw = pd.read_csv(
    raw_data_path,
    skiprows=3,
    header=None,
    names=['Date', 'Close'],
    parse_dates=['Date']
)
data_raw.set_index('Date', inplace=True)

# Load data sentimen
sentiment_df = pd.read_csv(sentiment_path, parse_dates=['Date'])

# Load data harian yang sudah diproses
data_daily = pd.read_csv(processed_features_path, parse_dates=['Date'], index_col='Date')

# ==================== B. Pratinjau Dataset Saham & Sentimen ====================
# Tanggal yang tersedia di dataset
available_dates = data_raw.index.date
tanggal_saham = [d for d in pd.date_range('2018-02-03', '2018-02-13', freq='D').date if d in available_dates]

# Tabel 1: Data saham mentah
preview_raw = data_raw.loc[tanggal_saham].sort_index()
preview_raw.to_csv(os.path.join(gambar_path, "preview_saham_mentah.csv"))
display_table(preview_raw.reset_index(), "Tabel 1: Pratinjau Data Saham Mentah")

# Tabel 2: Data sentimen
sentiment_preview = sentiment_df[sentiment_df['Date'].isin(tanggal_saham)].sort_values('Date')
sentiment_preview.to_csv(os.path.join(gambar_path, "preview_sentimen.csv"), index=False)
display_table(sentiment_preview, "Tabel 2: Pratinjau Data Sentimen")

# Tabel 3: Gabungan data saham dan sentimen
combined_preview = pd.merge(
    preview_raw.reset_index(),
    sentiment_preview,
    on='Date',
    how='left'
)
combined_preview['Sentiment'] = combined_preview['Sentiment'].fillna(0)
display_table(combined_preview[['Date', 'Close', 'Sentiment']], "Tabel 3: Gabungan Data Saham dan Sentimen")

# ==================== C. Resampling & Overlay Sentimen ====================
# Resampling & Interpolasi
plt.figure(figsize=(12, 5))
plt.plot(preview_raw.index, preview_raw['Close'], 'o', label='Original (with gaps)', alpha=0.6)
plt.plot(preview_raw.index, preview_raw['Close'], '-', label='Resampled + Interpolated')
plt.title("Resampling & Interpolasi Harga Saham")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(gambar_path, "resample_interpolasi.png"), dpi=300)

# Overlay Harga Saham dan Sentimen
fig, ax1 = plt.subplots(figsize=(12, 5))
ax2 = ax1.twinx()
ax1.plot(preview_raw.index, preview_raw['Close'], 'b-', label='Harga Saham')
ax2.plot(preview_raw.index, combined_preview['Sentiment'], 'r--', label='Sentimen')
ax1.set_ylabel('Harga Saham (USD)', color='b')
ax2.set_ylabel('Sentimen Komposit', color='r')
plt.title("Overlay Harga Saham dan Sentimen")
plt.tight_layout()
plt.savefig(os.path.join(gambar_path, "overlay_sentimen_saham.png"), dpi=300)

# ==================== D. MA, RSI, Volatilitas ====================
# Ambil data terbaru yang tersedia
end_date = data_daily.index.max()
start_date = end_date - pd.DateOffset(days=100)
recent_data = data_daily.loc[start_date:end_date]

plt.figure(figsize=(12, 5))
plt.plot(recent_data.index, recent_data['Close'], label='Harga Saham')
plt.plot(recent_data.index, recent_data['MA_5'], label='MA 5 Hari')
plt.plot(recent_data.index, recent_data['MA_20'], label='MA 20 Hari')
plt.title("Moving Average 5 & 20 Hari")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(gambar_path, "ma5_ma20.png"), dpi=300)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8), sharex=True)
ax1.plot(recent_data.index, recent_data['RSI_14'], color='purple')
ax1.axhline(70, color='r', linestyle='--', alpha=0.3)
ax1.axhline(30, color='g', linestyle='--', alpha=0.3)
ax1.set_title('RSI 14 Hari')
ax1.set_ylim(0, 100)
ax1.grid(True)

ax2.plot(recent_data.index, recent_data['Volatility_5D'], color='orange')
ax2.set_title('Volatilitas 5 Hari')
ax2.grid(True)

plt.tight_layout()
plt.savefig(os.path.join(gambar_path, "rsi_volatilitas.png"), dpi=300)

# ==================== Tabel 4: Preview Fitur Teknikal ====================
# Ambil 5 data pertama setelah dropna
preview_fitur = data_daily.iloc[:5].copy()
preview_fitur = preview_fitur[[
    'lag_1', 'lag_2', 'lag_3', 'lag_4', 'lag_5', 
    'MA_5', 'MA_20', 'Return_1D', 'Volatility_5D', 
    'RSI_14', 'Momentum_5D', 'MA_5_20_ratio', 'RSI_MA5'
]]
preview_fitur.to_csv(os.path.join(gambar_path, "preview_fitur_teknikal.csv"))
display_table(preview_fitur.reset_index(), "Tabel 4: Pratinjau Fitur Teknikal")

# ==================== F. Gantt Chart Periode COVID ====================
covid_periods = {
    'Sebelum COVID': ('2018-01-01', '2020-02-28'),
    'Selama COVID': ('2020-03-01', '2022-12-31'),
    'Setelah COVID': ('2023-01-01', '2024-01-01')
}

# Hitung jumlah data per periode
rows = []
for p, (start, end) in covid_periods.items():
    data_p = data_daily.loc[start:end]
    split = int(len(data_p) * 0.8)
    rows.append({
        'Periode': p,
        'Jumlah Train': split,
        'Jumlah Test': len(data_p) - split,
        'Total': len(data_p)
    })
df_split = pd.DataFrame(rows)
df_split.to_csv(os.path.join(gambar_path, "pembagian_data_periode.csv"), index=False)
display_table(df_split, "Tabel 5: Pembagian Data per Periode COVID")

# Visualisasi Gantt Chart
fig, ax = plt.subplots(figsize=(10, 3))
colors = ['green', 'orange', 'red']
for i, (label, (start, end)) in enumerate(covid_periods.items()):
    ax.barh(0, pd.to_datetime(end) - pd.to_datetime(start),
            left=pd.to_datetime(start), color=colors[i], label=label)
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.yticks([])
plt.legend()
plt.title("Pembagian Periode COVID untuk Data")
plt.tight_layout()
plt.savefig(os.path.join(gambar_path, "gantt_covid_periods.png"), dpi=300)

# ==================== Tabel 6: Preview Data per Periode ====================
# Fungsi untuk mencari contoh data dengan sentimen tertentu
def find_examples(df, values):
    examples = []
    for val in values:
        # Cari yang paling mendekati nilai yang diinginkan
        idx = (df['Sentiment'] - val).abs().idxmin()
        examples.append(df.loc[[idx]])
    return pd.concat(examples)

# Cari contoh data untuk setiap periode
covid_samples = []
for period, (start, end) in covid_periods.items():
    period_data = data_daily.loc[start:end]
    examples = find_examples(period_data, [0, 1, -1])
    examples['Periode'] = period
    covid_samples.append(examples)

covid_samples_df = pd.concat(covid_samples)
covid_samples_df = covid_samples_df[[
    'Close', 'Sentiment', 'Sentiment_Lag1', 'lag_1', 'lag_2', 'lag_3', 
    'lag_4', 'lag_5', 'MA_5', 'MA_20', 'Return_1D', 'Volatility_5D', 
    'RSI_14', 'Momentum_5D', 'MA_5_20_ratio', 'RSI_MA5', 'Periode'
]]
covid_samples_df.to_csv(os.path.join(gambar_path, "preview_data_periode.csv"))
display_table(covid_samples_df.reset_index(), "Tabel 6: Contoh Data per Periode COVID")

# ==================== Tabel 7: Preview Dataset Akhir ====================
# Gunakan tanggal yang sama dengan Tabel 1
preview_akhir = data_daily.loc[tanggal_saham].head(5)
preview_akhir = preview_akhir[[
    'Close', 'Sentiment', 'Sentiment_Lag1', 'lag_1', 'lag_2', 'lag_3', 
    'lag_4', 'lag_5', 'MA_5', 'MA_20', 'Return_1D', 'Volatility_5D', 
    'RSI_14', 'Momentum_5D', 'MA_5_20_ratio', 'RSI_MA5'
]]
preview_akhir.to_csv(os.path.join(gambar_path, "preview_dataset_akhir.csv"))
display_table(preview_akhir.reset_index(), "Tabel 7: Pratinjau Dataset Akhir")

# ==================== G. Heatmap Korelasi ====================
plt.figure(figsize=(12, 10))
corr = data_daily[[
    'Close', 'Sentiment', 'Sentiment_Lag1', 'MA_5', 'MA_20', 
    'Return_1D', 'Volatility_5D', 'RSI_14', 'Momentum_5D',
    'MA_5_20_ratio', 'RSI_MA5', 'lag_1', 'lag_2', 'lag_3', 'lag_4', 'lag_5'
]].corr()

# Format korelasi untuk ditampilkan
corr_display = corr.copy()
for col in corr_display.columns:
    corr_display[col] = corr_display[col].apply(lambda x: f"{x:.2f}")

mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, cmap='coolwarm', fmt=".2f", 
            annot_kws={"size": 8}, cbar_kws={"shrink": 0.8})
plt.title("Heatmap Korelasi Antar Fitur", fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.yticks(fontsize=8)
plt.tight_layout()
plt.savefig(os.path.join(gambar_path, "heatmap_korelasi.png"), dpi=300)

# Tampilkan tabel korelasi
display_table(corr_display, "Tabel Korelasi Antar Fitur")

# ==================== H. Visualisasi Prediksi vs Aktual ====================
def load_prediction_data(base_path, period_name):
    """Muat data prediksi dari semua run untuk periode tertentu"""
    all_predictions = []
    run_dirs = [d for d in os.listdir(base_path) if d.startswith('run_')]
    
    for run_dir in run_dirs:
        file_name = f"predictions_{period_name.lower().replace(' ', '_')}.csv"
        file_path = os.path.join(base_path, run_dir, file_name)
        
        if os.path.exists(file_path):
            df = pd.read_csv(file_path, parse_dates=['Date'])
            df['Run'] = run_dir
            all_predictions.append(df)
    
    if not all_predictions:
        return None
    
    return pd.concat(all_predictions)

def plot_avg_predictions(base_path, model_name):
    """Buat plot prediksi rata-rata vs aktual untuk setiap periode"""
    periods = {
        'Sebelum COVID': ('2018-01-01', '2020-02-29'),
        'Selama COVID': ('2020-03-01', '2022-12-31'),
        'Setelah COVID': ('2023-01-01', '2024-01-01'),
        'Keseluruhan': ('2018-01-01', '2024-01-01')
    }
    
    for period_name, (start_date, end_date) in periods.items():
        # Muat data prediksi
        pred_df = load_prediction_data(base_path, period_name)
        if pred_df is None or pred_df.empty:
            print(f"⚠️ Tidak ada data prediksi untuk periode {period_name}")
            continue
        
        # Filter data aktual untuk periode yang sesuai
        actual_data = data_daily.loc[start_date:end_date, ['Close']].reset_index()
        actual_data = actual_data.rename(columns={'Close': 'Actual'})
        
        # Gabungkan semua prediksi
        merged = pd.merge(actual_data, pred_df, on='Date', how='inner')
        
        # Hitung rata-rata prediksi dari semua run
        avg_pred = merged.groupby('Date').agg({
            'Actual': 'first',
            'Predicted': 'mean'
        }).reset_index()
        
        # Buat plot
        plt.figure(figsize=(12, 6))
        plt.plot(avg_pred['Date'], avg_pred['Actual'], 'b-', label='Aktual', linewidth=2)
        plt.plot(avg_pred['Date'], avg_pred['Predicted'], 'r--', label='Prediksi Rata-rata', linewidth=1.5)
        
        # Format tanggal
        plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
        plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=2))
        plt.gcf().autofmt_xdate()
        
        plt.title(f"ARIMA: Prediksi vs Aktual ({period_name})")
        plt.xlabel('Tanggal')
        plt.ylabel('Harga Saham (USD)')
        plt.legend()
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.tight_layout()
        
        # Simpan plot
        file_name = f"prediksi_vs_aktual_{period_name.replace(' ', '_').lower()}.png"
        plt.savefig(os.path.join(gambar_path, file_name), dpi=300)
        plt.close()
        print(f"✅ Plot prediksi vs aktual untuk {period_name} disimpan")

# Panggil fungsi untuk membuat plot prediksi vs aktual
plot_avg_predictions(base_path, "ARIMA")

print("\n✅ Semua visualisasi dan pratinjau tabel ARIMA berhasil disimpan di folder:", gambar_path)
```
